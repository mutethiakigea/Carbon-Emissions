<h1 style="text-align: center; color: Orange;">Carbon Emissions</h1>

+ Carbon emissions are the discharge of carbon dioxide gas into the atmosphere. Carbon dioxide is a greenhouse gas that plays a significant role in climate change and global warming. Burning fossil fuels for transportation and energy production is one activity that contributes significantly to these emissions (Driga & Drigas, 2019).
+ Carbon dioxide is a byproduct of burning fossil fuels in industries, vehicles, and other stationary energy sources. Deforestation and different types of land use change can also release carbon from soil and trees into the sky (Koga et al.,2020). These emissions trap heat in the Earth's atmosphere, raising global temperatures and accelerating climate change.

<h2 style="text-align: center; color: Green;">Global Warming</h2>

* Global warming refers to the long-term increase in Earth's average surface temperature due to the buildup of greenhouse gases in the atmosphere. This phenomenon is a key aspect of climate change. Greenhouse gases, such as carbon dioxide (CO2), methane (CH4), and water vapor (H2O), naturally exist in the Earth's atmosphere and have the ability to trap heat from the sun (Aliyu et al., 2020). This natural greenhouse effect is essential for maintaining temperatures on Earth that are conducive to life.

* However, human activities, particularly the burning of fossil fuels (such as coal, oil, and natural gas) and deforestation, have significantly increased the concentration of greenhouse gases in the atmosphere (Shivanna, 2022). This enhanced greenhouse effect has led to an excess of heat being trapped in the atmosphere, causing the Earth's average temperature to rise. This rise in temperature is known as global warming.

### Carbon Emission Forecast in the next 5 years 

```python
from sklearn.linear_model import LinearRegression
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
import statsmodels.api as sm
```

```python
# Select the "World" row for the years 1990-2018
world_emissions = emissions1[emissions1["Country"] == "World"].iloc[:, 6:-1].values.flatten()
```

```python
# Create a time series object
time_series = pd.Series(world_emissions, index=pd.date_range(start="1990", periods=len(world_emissions), freq='Y'))
```

```python
# Define the forecasting period (next 5 years from 2018)
forecast_period = 5
forecast_dates = pd.date_range(start="2019", periods=forecast_period, freq='Y')
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
# Fit an ARIMA time series forecasting model
# You may need to adjust the model parameters based on your data
model = ARIMA(time_series, order=(0, 1, 1))  # Example order, adjust as needed
results = model.fit()
```

```python
# Forecast future emissions
forecast = results.forecast(steps=forecast_period)
```

```python
# Forecast future emissions
forecast, stderr, conf_int = results.forecast(steps=forecast_period)
```

```python
# Visualize the historical data and forecast
plt.figure(figsize=(12, 6))
plt.plot(time_series, label='Historical Data', marker='o')
plt.plot(forecast_dates, forecast, label='Forecast', linestyle='--', marker='o', color='orange')
plt.xlabel('Year')
plt.ylabel('Emissions')
plt.title('Carbon Emission Forecast for "World" (2019-2023)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Print the forecasted values for the next 5 years
print("Forecasted Emissions for the next 5 years (2019-2023):")
print(forecast)
```

The forecasted emissions for the "World" for the next five years (2019-2023) indicate a decrease from the emissions 2018. That suggests a declining trend in carbon emissions during this period. Here are the forecasted emissions for each year:

<h1 style="text-align: center; color: Green;">Factors that may have led to a decrease in Carbon emission</h1>

A decrease in carbon emissions can be attributed to various factors, as efforts are made to reduce greenhouse gas emissions and combat climate change. Here are five key factors that may have contributed to a decrease in carbon emissions:
1. Shift to Renewable Energy: A transition from fossil fuels to renewable energy sources such as wind, solar, and hydropower reduces the carbon footprint of energy generation(Bayazit, 2021). Investments in clean energy technologies and policies promoting their adoption have played a significant role.
2. Energy Efficiency: Improved energy efficiency in industries, transportation, and buildings reduces the energy required to accomplish tasks, resulting in lower emissions. That includes the adoption of energy-efficient technologies and practices.
3. Policy and Regulations: Government policies and regulations, such as carbon pricing, emissions standards, and renewable energy targets, can incentivize businesses and individuals to reduce emissions. International agreements like the Paris Agreement have also contributed to global emissions reduction efforts.
4. Technological Advances: Technology advancements have led to the development of cleaner and more fuel-efficient vehicles and innovations in carbon capture and storage (CCS) and sustainable agriculture practices (Xin et al., 2023). These technologies help mitigate emissions across various sectors.
5. Awareness and Behavior Change: Growing awareness of the environmental impact of carbon emissions has led to changes in consumer behaviour and lifestyle choices. That includes increased use of public transportation, reduced meat consumption (associated with high emissions), and a shift toward eco-friendly products.
